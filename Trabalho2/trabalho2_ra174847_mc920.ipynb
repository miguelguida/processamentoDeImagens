```python
# TRABALHO 2 DE MC920

# ALUNO: Miguel Augusto Silva Guida   RA: 174847


import cv2
import numpy as np
import matplotlib.pyplot as plt



imgcv = cv2.imread('butterfly.png', cv2.IMREAD_GRAYSCALE)

# Criacao das mascaras.
h1 = np.array([[0, 0, -1, 0, 0], [0, -1, -2, -1, 0], [-1, -2, 16, -2, -1],[0, -1, -2, -1, 0], [0, 0, -1, 0, 0]])
h2 = np.array([[1, 4, 6, 4, 1], [4, 16, 24, 16, 4], [6, 24, 36, 24, 6], [4, 16, 24, 16, 4], [1, 4, 6, 4, 1]])
h3 = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])
h4 = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]])
h5 = np.array([[-1, -1, -1], [-1, 8, -1], [-1, -1, -1]])
h6 = np.array([[1, 1, 1], [1, 1, 1], [1, 1, 1]])
h7 = np.array([[-1, -1, 2], [-1, 2, -1], [2, -1, -1]])
h8 = np.array([[2, -1, -1], [-1, 2, -1], [-1, -1, 2]])


h2 = h2/256
h6 = h6/9

#Rotaciona mascaras em 180 graus para aplicar a convolucao.
h1 = np.flip(h1)
h2 = np.flip(h2)
h3 = np.flip(h3)
h4 = np.flip(h4)
h5 = np.flip(h5)
h6 = np.flip(h6)
h7 = np.flip(h7)
h8 = np.flip(h8)

# Aplica a convolucao das mascaras na imagem. Armazena as imagens resultantes nas variaveis.
cc1 = cv2.filter2D(imgcv,-1, h1)
cc2 = cv2.filter2D(imgcv,-1, h2)
cc3 = cv2.filter2D(imgcv,-1, h3)
cc4 = cv2.filter2D(imgcv,-1, h4)
cc5 = cv2.filter2D(imgcv,-1, h5)
cc6 = cv2.filter2D(imgcv,-1, h6)
cc7 = cv2.filter2D(imgcv,-1, h7)
cc8 = cv2.filter2D(imgcv,-1, h8)


# Operacoes nas imagens resultantes de h3 e h4 para formar o h9.
ratio3 = 1/np.max(cc3) 
a3 = np.max(cc3)     
ratio4 = 1/np.max(cc4) 
a4 = np.max(cc4)     
cc33 = np.floor(np.multiply(cc3*ratio3,cc3*ratio3)* a3)
cc44 = np.floor(np.multiply(cc4*ratio4,cc4*ratio4)* a4)

# Calcura a raiz da soma dos quadrados.
cc9 = np.floor(np.sqrt(cc33+cc44))

# Normaliza a imagem.
cc9 = np.floor(cc9*(255/np.max(cc9)))


plt.imshow(cc1, cmap='gray')
plt.show()
plt.imshow(cc2, cmap='gray')
plt.show()
plt.imshow(cc3, cmap='gray')
plt.show()
plt.imshow(cc4, cmap='gray')
plt.show()
plt.imshow(cc5, cmap='gray')
plt.show()
plt.imshow(cc6, cmap='gray')
plt.show()
plt.imshow(cc7, cmap='gray')
plt.show()
plt.imshow(cc8, cmap='gray')
plt.show()

cv2.imwrite('butterflyh1.png', cc1)
cv2.imwrite('butterflyh2.png', cc2)
cv2.imwrite('butterflyh3.png', cc3)
cv2.imwrite('butterflyh4.png', cc4)
cv2.imwrite('butterflyh5.png', cc5)
cv2.imwrite('butterflyh6.png', cc6)
cv2.imwrite('butterflyh7.png', cc7)
cv2.imwrite('butterflyh8.png', cc8)
cv2.imwrite('butterflyh9.png', cc9)




```
