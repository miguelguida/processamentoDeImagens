```python
def reduce_resolution_in_half(img):
    dimension = int(len(img)/2)
    r = np.zeros((dimension,dimension,4))
    r[::,::,0] = img[::2, ::2]
    r[::,::,1] = img[1::2, ::2]
    r[::,::,2] = img[::2, 1::2]
    r[::,::,3] = img[1::2, 1::2]
    resized = np.floor(np.mean(r, axis=2))
    
    img[::2, ::2] = resized
    img[1::2, ::2] = resized
    img[::2, 1::2] = resized
    img[1::2, 1::2] = resized
    
    return img
    
```
