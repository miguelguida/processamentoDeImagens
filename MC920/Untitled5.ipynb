```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('baboon.png', cv2.IMREAD_GRAYSCALE)

dimension = int(len(img)/2)
r = np.zeros((dimension,dimension,4))
r[::,::,0] = img[::2, ::2]
r[::,::,1] = img[1::2, ::2]
r[::,::,2] = img[::2, 1::2]
r[::,::,3] = img[1::2, 1::2]


#resized = reduce_resolution_in_half(img)
def reduce_resolution_in_half(img):
    dimension = int(len(img)/2)
    r = np.zeros((dimension,dimension,4))
    r[::,::,0] = img[::2, ::2]
    r[::,::,1] = img[1::2, ::2]
    r[::,::,2] = img[::2, 1::2]
    r[::,::,3] = img[1::2, 1::2]
    resized = np.floor(np.mean(r, axis=2))
    
    return resized

def reduce_resolution_in_half_vectorized(img):
    dimension =  int(len(img)/2)
    print(dimension)
    resized = cv2.resize(img,dsize=(dimension,dimension), interpolation=cv2.INTER_AREA)
    return resized


resized = reduce_resolution_in_half(img)
saved = cv2.resize(resized,dsize=(512,512), interpolation=cv2.INTER_AREA)
cv2.imwrite('baboon256.png', saved)
plt.imshow(resized, cmap='gray')
plt.show()
resized = reduce_resolution_in_half(resized)
saved = cv2.resize(resized,dsize=(512,512), interpolation=cv2.INTER_AREA)
cv2.imwrite('baboon128.png', saved)
plt.imshow(resized, cmap='gray')
plt.show()
resized = reduce_resolution_in_half(resized)
saved = cv2.resize(resized,dsize=(512,512), interpolation=cv2.INTER_AREA)
cv2.imwrite('baboon64.png', saved)
plt.imshow(resized, cmap='gray')
plt.show()
resized = reduce_resolution_in_half(resized)
saved = cv2.resize(resized,dsize=(512,512), interpolation=cv2.INTER_AREA)
cv2.imwrite('baboon32.png', saved)
plt.imshow(resized, cmap='gray')
plt.show()
resized = reduce_resolution_in_half(resized)
saved = cv2.resize(resized,dsize=(512,512), interpolation=cv2.INTER_AREA)
cv2.imwrite('baboon16.png', saved)
plt.imshow(resized, cmap='gray')
plt.show()
resized = reduce_resolution_in_half(resized)
saved = cv2.resize(resized,dsize=(512,512), interpolation=cv2.INTER_AREA)
cv2.imwrite('baboon8.png', saved)
plt.imshow(resized, cmap='gray')
plt.show()

```
